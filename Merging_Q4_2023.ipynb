```python
import pandas as pd
```

---

Import clean data and view head

```python
oct2023 = pd.read_excel(r"D:\Data Scraping\Turo\Nashville\Octorber\Clean Data\clean_data.xlsx")
jan2024 = pd.read_excel(r"D:\Data Scraping\Turo\Nashville\Jan 2024\clean data\clean data.xlsx")
```

```python
jan2024.head()
```

---

Merge oct2023 with jan2024 data

```python
q4_data = pd.merge(oct2023, jan2024, on = 'id', how = 'outer')

q4_data.head()
```

---

Create a column for Q4_2023 trips

```python
q4_data['trips_Q4_2023'] = q4_data['trips_jan2024'] - q4_data['trips_oct2023']

q4_data.head()
```

---

Create a column for Q4_2023 price increase

```python
q4_data['price_increase_Q4_2023'] = q4_data['average_price_jan2024'] - q4_data['average_price_oct2023']

q4_data.head()
```

---

Create a column for Q4_2023 price increase

```python
q4_data['rating_increase_Q4_2023'] = q4_data['average_rating_jan2024'] - q4_data['average_rating_oct2023']

q4_data.head()
```

---

export to excel

```python
q4_data.to_excel(r"D:\Data Scraping\Turo\Nashville\clean data\q4_2023_data.xlsx", index = False)
```
