```python
import pandas as pd
import os
```

Import the entire folder and create a list of file paths for the individual csv files

```python
#Innitiate an empty list to store the csv file paths
folder_path = r"D:\Data Scraping\Turo\Nashville\Jan 2024\excel files"
file_paths = []

# List all files in the folder (excluding subdirectories)
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        file_paths.append(file_path)

# View the file paths
file_paths
```

---

Concatenating data frames

```python
#Innitiate an empty list to store the data frames
cars_dfs = []

#Use the file paths to fill the list with data frames
for df in file_paths: 
    df = pd.read_csv(df)
    cars_dfs.append(df)

#Concatenate the data frames together
concat_df = pd.concat(cars_dfs)
concat_df
```

---

Filter columns

```python
#Sub-set only the columns of interest
concat_df = concat_df[['vehicles.id', 
                      'vehicles.completedTrips',
                      'vehicles.hostId',
                      'vehicles.isAllStarHost',
                      'vehicles.isFavoritedBySearcher', 
                      'vehicles.isNewListing',
                      'vehicles.make',
                      'vehicles.model',
                      'vehicles.rating',
                      'vehicles.type',
                      'vehicles.year',
                      'vehicles.avgDailyPrice.amount',
                      'vehicles.location.city',
                      'vehicles.location.homeLocation.lat',
                      'vehicles.location.homeLocation.lng']]

concat_df
```

---

Rename columns

```python
#Create new column names
new_column_names = {'vehicles.id':'id', 
                      'vehicles.completedTrips':'trips_jan2024',
                      'vehicles.hostId':'host_id_jan2024',
                      'vehicles.isAllStarHost':'all_star_host_jan2024',
                      'vehicles.isFavoritedBySearcher':'favorited_by_searcher_jan2024', 
                      'vehicles.isNewListing':'new_listing_jan2024',
                      'vehicles.make':'make_jan2024',
                      'vehicles.model':'model_jan2024',
                      'vehicles.rating':'rating_jan2024',
                      'vehicles.type':'type_jan2024',
                      'vehicles.year':'year_jan2024',
                      'vehicles.avgDailyPrice.amount':'price_per_day_jan2024',
                      'vehicles.location.city':'city_jan2024',
                      'vehicles.location.homeLocation.lat':'lat_jan2024',
                      'vehicles.location.homeLocation.lng':'lng_jan2024'}

#Apply new column names
new_df = concat_df.rename(columns=new_column_names)
new_df
```

---

Remove duplicates

```python
unique_data = new_df.drop_duplicates()

unique_data
```

---

Find the total amount of unique values in the 'id' column

```python
unique_data[['id']].drop_duplicates()
```

---

Average the price

```python
#Group by id and return a df with averge price for each id
join1 = unique_data.groupby('id')['price_per_day_jan2024'].agg(average_price='mean').reset_index()

#Remove the price column from the original df and drop the duplicates
join2 = unique_data.drop(columns = ['price_per_day_jan2024']).drop_duplicates()

#Join the average price df to the original df
final_clean = pd.merge(join1, join2, how = 'inner', on = 'id')

final_clean = final_clean.rename(columns = {'average_price' : 'average_price_jan2024'})

final_clean
```

---

Average the rating

```python
#Group by id and return a df with averge trip rating for each id
join1 = final_clean.groupby('id')['rating_jan2024'].agg(average_rating='mean').reset_index()

#Remove the rating column from the original df and drop the duplicates
join2 = final_clean.drop(columns = ['rating_jan2024']).drop_duplicates()

#Join the average rating df to the original df
final_clean = pd.merge(join1, join2, how = 'inner', on = 'id')

final_clean = final_clean.rename(columns = {'average_rating' : 'average_rating_jan2024'})

final_clean
```

---

If one car has two different values for 'trips', then take the larger one

```python
#Group by id and return a df with max trips for each id
join1 = final_clean.groupby('id')['trips_jan2024'].max().reset_index()

#Remove the trips column from the original df and drop the duplicates
join2 = final_clean.drop(columns = ['trips_jan2024']).drop_duplicates()

#Join the trips df to the original df
final_clean = pd.merge(join1, join2, how = 'inner', on = 'id')

final_clean = final_clean.rename(columns = {'trips' : 'trips_jan2024'})

final_clean
```

---

Filter for only cities of interest

```python
final_clean['city_jan2024'].unique()
```

```python
#Create a list of cities to keep in the dataset
cities_to_keep = ['Nashville', 'Brentwood', 'Murfreesboro',
       'Hendersonville', 'Franklin', 'Smyrna', 'Goodlettsville',
       'Hermitage', 'Antioch', 'Mt. Juliet',
       'La Vergne', 'Mount Juliet', 'Forest Hills', 'Nolensville', 'Arrington', 'Lavergne',
       'Mt Juliet', 'Whites Creek']

#Filter for only the cities of interest
final_clean = final_clean[final_clean['city_jan2024'].isin(cities_to_keep)]
```

```python
final_clean['city_jan2024'].unique()
```

---

Rename columns 

```python
#Create a dictionary with new names
value_mapping = {'Mt. Juliet':'Mount Juliet',
                 'Mt Juliet':'Mount Juliet',
                 'Forest Hills':'Nashville',
                'Antioch':'Nashville',
                'Whites Creek':'Nashville',
                'Hermitage':'Nashville',
                'Lavergne':'La Vergne'}

#Apply the new names to the city column
final_clean.loc[final_clean['city_jan2024'].isin(value_mapping.keys()), 'city_jan2024'] = final_clean['city_jan2024'].replace(value_mapping)
```
